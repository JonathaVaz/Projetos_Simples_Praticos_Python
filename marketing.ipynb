```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
planos = ["Pano Individual", "Plano Casal", "Plano Família"]
valores = [80, 140, 270]

plt.style.use("dark_background")

plt.figure(figsize=(12,10))
plt.title("Tabela de assinatura de planos quando possui 3 ou mais pessoas por assinatura\n", fontsize=18, fontweight='bold', style='italic')
plt.bar(planos, valores, color='purple', edgecolor='white')
plt.xticks(fontsize=11)
plt.yticks(fontsize=12)
plt.ylabel("\nPreços\n", fontsize=14, fontweight='bold', style='italic')
plt.xlabel('\nPlanos', fontsize=14, fontweight='bold', style='italic')
for x,y in zip(planos,valores):
    plt.text(x, y+0.05, '%d' % y, ha='center' , va= 'bottom', fontsize=14, fontweight='bold', style='italic')
```

```python
precos = [70, 130, 250]
planos = ['Individual', 'Casal', 'Família']

paleta = sns.color_palette('gnuplot2')
cores = sns.color=('teal', 'mediumspringgreen', 'turquoise')

plt.style.use("dark_background")

plt.figure(figsize=(12,10))
plt.title("Valores dos Planos Disponiveis\n", fontsize=18, fontweight='bold', style='italic')
plt.bar(planos, precos, color=paleta, edgecolor='white')
plt.xticks(fontsize=11)
plt.yticks(fontsize=12)
plt.ylabel("\nPreços\n", fontsize=14, fontweight='bold', style='italic')
plt.xlabel('\nPlanos', fontsize=14, fontweight='bold', style='italic')
for x,y in zip(planos,precos):
    plt.text(x, y+3, '%d' % y, ha='center' , va= 'bottom', fontsize=14, fontweight='bold', style='italic')
```
